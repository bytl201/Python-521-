```python
def display_board(board):
    print("--------------") # show 14 dashes here
    for row in board:
        print("|", " | ".join(row), "|")
        print("--------------") # show 14 dashes here
```

```python
def player_input(board: list, current_player: str):
    while(True):
        row_num = int(input("Row Number"))-1
        col_num = int(input("Column Number"))-1

        if board[row_num][col_num] in ['X','O']:
            print("This location is taken. Please enter a new location.")
        else:
            board[row_num][col_num] = current_player
            display_board(board)
            break
```

```python
def check_win(board: list, current_player: str):
    list_to_check = [current_player, current_player, current_player]

    for row in board:
        if row == list_to_check:
            return True
        
    for col in range(3):
        temp_list = [board[row][col] for row in range(3)]
        if temp_list == list_to_check:
            return True
    
        
    if [board[i][i] for i in range(len(board))] == list_to_check:
        return True
    elif [board[i][2-i] for i in range(len(board))] == list_to_check:
        return True
```

```python
def check_tie(board):
    for row in range(len(board)):
        for col in range(len(board[row])):
            if board[row][col] not in ['X', 'O']:
                return False
    return True
```

```python
def switch_player(current_player: str):
    return 'X' if current_player == 'O' else 'O'
```

```python
import IPython.display 

print("Let's play tic-tac-toe!")

# Example tic-tac-toe board
tic_tac_toe_board = [[" ", " ", " "],
                    [" ", " ", " "],
                    [" ", " ", " "]]
display_board(tic_tac_toe_board)

current_player = 'X'

game_won = False
game_drawn = False

with open('scores_tracker.txt', mode = "r") as fo:
    scores = fo.read()

scores_dict = eval(scores)
print(f"The current high scores are: {scores_dict}")

while (not game_won) and (not game_drawn):

    IPython.display.clear_output
    player_input(tic_tac_toe_board, current_player)

    if check_win(tic_tac_toe_board, current_player):
        print(f"{current_player} has won!")
        game_won = True
        scores_dict[current_player] += 1
    elif check_tie(tic_tac_toe_board):
        print(f"The game is tied!")
        game_drawn = True
        scores_dict['Tied'] += 1
    else:
        current_player = switch_player(current_player)

with open('scores_tracker.txt', mode="w") as file_object:
    high_scores_string = str(scores_dict)
    file_object.write(high_scores_string)
```
