## 1. Pascal's Triangle

## Part A and B

```python
def make_new_row(old_row: list):
    '''
    This function creates the next row by taking current row in the pascal's triangle and using the 'old_row' to construct the next row in the triangle.
    '''

    # edge case for empty 'old_row'
    if not old_row:
        # return [1] as new row
        return [1]

    # start with the new row having '1' at both ends and the first element is always 1
    new_row = [1]  
    
    # iterate the len of 'old_row' and if len of 'old_row' is 1, the for loop won't start so it will skip to code below for loop
    for i in range(1, len(old_row)):
        # compute the middle elements by summing adjacent elements in the 'old row'
        new_row.append(old_row[i - 1] + old_row[i])
    
    # add the last element as 1 (since it's always 1 at the edge)
    new_row.append(1)

    # return the new_row
    return new_row

row_0 = make_new_row([])
print(row_0)
row_1 = make_new_row([1])
print(row_1)
row_2 = make_new_row([1,1])
print(row_2)
```

## Part C

```python
def pascals_triangle(height: int):
    '''
    This function creates pascals triangle by taking in desired height for the triangle.
    '''

    # starter triangle
    triangle = [[1]]  
    
    # iterate from index 1 to heigh
    for _ in range(1, height):
        # print(_)
        # print(triangle[-1])

        # generate the next row based on the last list in 'triangle'
        new_row = make_new_row(triangle[-1])

        # print(new_row)

        # append the 'new_row' to 'triangle' list
        triangle.append(new_row)  

        # print(triangle)
        # print()
    
    return triangle

# ask user for 'height' input
height = int(input("Enter the desired height of Pascal's triangle: "))

# create pascals triangle
triangle = pascals_triangle(height)

# print each row in the triangle
for row in triangle:
    print(row)
```

## Part D

```python
def print_pascals_triangle(triangle):
    '''
    This function cleanly outputs the pascal's triangle centered and as strings instead of lists.
    '''
    # find the width of the last row in the 'triangle'
    max_width = len(' '.join(map(str, triangle[-1])))  # Get the length of the last row in string format

    for row in triangle:
        # join the numbers in the row into a string with spaces, and then center it
        row_str = ' '.join(map(str, row))

        # print(row_str)

        # center the row within the max width
        print(row_str.center(max_width))  

# ask user for 'height' input
height = int(input("Enter the desired height of Pascal's triangle: "))

# create pascals triangle
triangle = pascals_triangle(height)

# output as centered characters
print_pascals_triangle(triangle)
```

## 2. DNA Sequencing

## Part A

```python
def find(full_string: str, substring: str, start: int, end: int):
    '''
    This function behaves like .find(), that returns the lowest index of the substring found in the 'full_string'.
    '''

    # validate that start and end indexes are valid
    if start < 0:
        start = 0
    if end > len(full_string):
        end = len(full_string)

    # search from start index to valid ending positions
    for i in range(start, end- len(substring) + 1):

        # check from the index to len(substring) if it matches the substring
        if full_string[i:i+len(substring)] == substring:
            # return the index start
            return i
    # return -1 if substring not found
    return -1

find("hello", "ell", 0, 6)
    
```

## Part B

```python
def multi_find(full_string: str, substring: str, start: int, end: int):
    '''
    This function returns all indexes in a string separated with commas of where the 'substring' was found.
    '''

    # hold all of indexes
    indices = []

    # retrieve the first index with substring
    pos = find(full_string, substring, start, end)

    # if 'pos' returned then add one to 'pos' for starting position and find other index with same 'substring'
    while pos != -1:
        # add index to list
        indices.append(str(pos))
        # move starting position up 1 and find next occurrence
        pos = find(full_string, substring, pos + 1, end)

    # return as a string with commas between each index
    return ",".join(indices)

print(multi_find("hello", "l", 0, 5))       # "2,3"
print(multi_find("banana", "ana", 0, 6))    # "1,3"
print(multi_find("abcdef", "gh", 0, 6) == "")     # "" (True)
```
