## 1. Countdown using recursion

```python
def count_down(num: int):
    """
    Recursively prints a countdown pattern of stars,
    decreasing from `num` down to 1.
    """

    # print "*" num times
    print("*" * num)
    # continue the countdown recursively until num reaches 0
    return None if num <= 0 else count_down(num-1)

# get number from user
num_times = int(input("Enter the countdown limit: "))
# initiate the function
count_down(num_times)
print("Blast Off!")
```

## 2. Sum of series using recursive functions

```python
def sum_series(i: int):
    """
    Recursively call sum_series() to calculate sum of series.
    """

    # base case
    if i == 1:
        return 1 / 3
    
    # continue calculating the sum by subtracting 1
    return i / (i + 2) + sum_series(i - 1)


# user input
i = int(input("Enter i: "))
result = sum_series(i)

# output
print(f"m({i}) = {result}")
```

## 3. Longest word in a sentence

```python
def longest_word(string_list: list, longest_string: str = ""):
    """
    Recursively finds and returns the longest word in a list.
    """

    # base case, return the longest word when list is empty
    if not string_list:
        return longest_string
    
    # retrieve the last element in the list and remove from list
    last_word = string_list.pop()

    # check if the last element in the list is greater than the current longest word
    if len(last_word) > len(longest_string):
        # replace with the new longest word
        longest_string = last_word

    # recursively call until the list is finished
    return longest_word(string_list, longest_string)

# user input
input_str = input("Enter the sentence here: ")

# split whitespaces and convert to list
input_list = input_str.split()

# call function first time
word = longest_word(input_list)

# output to user
print(f"The longest word in the sentence entered is: {word}")
```
