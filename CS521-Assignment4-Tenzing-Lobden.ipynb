### 1. List Comprehension

```python
# 1.26
# 26a

# initialize list
list_of_even_numbers_for_loop = []
for i in range(2, 41, 2):
    # add to list
    list_of_even_numbers_for_loop.append(i)
print(f"26a. Output of creating a list of 20 even numbers without using if statement with regular for loop = {list_of_even_numbers_for_loop}\n")

# 26b
list_of_even_numbers_list_comprehension = [i for i in range(2, 41, 2)]
print(f"26b. Output of creating a list of 20 even numbers without using if statement with list comprehension = {list_of_even_numbers_list_comprehension}\n")

# 1.38
# 38

# get number from user
num = int(input("Enter a number:"))
# iterate from 1 to num+1 to include the user's number and check if user's num % 2 == 0
# if it's 0 then add to 'output' list
output = [i for i in range(1,num+1) if num % i == 0]
# use sum() to get the sum of all the numbers in the list
print(f"38. Output = {sum(output)}, List of factors = {output}\n")

# 1.63
# 63
# initialize the numbers
digits = [2,3,4,5]
digits1 = [9,9,1,9]
# multiply each digit by the power of 10 based on its position
# example: 2345, digits[i] = 2, len(digits) = 4, 2*10**(4-0-1) = 2*10**3 = 2*1000 = 2000
#          2345, digits[i] = 3, len(digits) = 4, 3*10**(4-1-1) = 3*10**2 = 3*100 = 300
number = [digits[i] * 10**(len(digits)-i-1) for i in range(len(digits))]
number1 = [digits1[i] * 10**(len(digits1)-i-1) for i in range(len(digits1))]

print(f"63. Output = {sum(number)}, List of number = {number}")
print(f"Output = {sum(number1)}, List of number1 = {number1}")
```

### 2. List Comprehension with Multiple Conditions

```python
# check if i%3 = 0 which means it is divisible by 3
# or 
# check if i%10 == 7 which means check if the last digit of i is 7. example: 7% 10 = 7; 27%10 = 7; 
output_list = [i for i in range(1,1001) if (i % 3 == 0) or (i % 10 == 7)]
print(f"2. Output of all the numbers from 1 to 1000 that are either divisible by 3 or end with digit 7:\n{output_list}\n")
```

### 3. Comprehension with else statements

```python
# 3a
# retrive list of numbers and then split up with "," (comma) being the delimiter to create an array of strings of number
input_list = input("Enter a list of numbers: ").split(",")
# convert each string in list to an int
input_list = [int(i) for i in input_list]
# create output with non-positives with 0 and then output positive numbers regularly
positive_num_list = [i if i > 0 else 0 for i in input_list]
print(f"3a. Output = {positive_num_list}\n")

# 3b
# use ternary operator with list comprehension to output "even" or "odd" in output_list
even_odd_list = ["even" if i % 2 == 0 else "odd" for i in input_list]
print(f"3b. Output = {even_odd_list}")
```
