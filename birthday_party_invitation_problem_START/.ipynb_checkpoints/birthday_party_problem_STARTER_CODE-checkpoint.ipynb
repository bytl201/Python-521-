## Useful methods
    
* Hint1: The Python File `readlines()` method https://www.w3schools.com/python/ref_file_readlines.asp
  - This method returns a list containing each **line** in the file as a list item.
* Hint2: Python String `replace()` method: https://www.w3schools.com/python/ref_string_replace.asp
  - As the name suggests, this method replaces a specified phrase with another specified phrase.
  - For example:
    ```python
    my_string = "Hello Avi, how's the weather today?"
    print(my_string)
    my_new_string = my_string.replace("Avi", "Avi Mohan")
    print(my_new_string)
    ```
* Hint3: Python String `strip()` method https://www.w3schools.com/python/ref_string_strip.asp
  - This method removes spaces at the beginning and end of a string.
  - Remember that `\n` looks like a space in text documents.

## Tasks and subtasks
Let's break the problem down into tasks (feel free to break tasks down into smaller subtasks if necessary).

1. Check step: Read the file `starting_letter.txt` and print its contents.
   - Make sure you enter both file name and path correctly!
   - If you choose not to use the `with` keyword, make sure to <font color='red'>close</font> the file once you've read its contents.


2. Use the `readlines()` method associated with `File` objects to split the contents of `starting_letter.txt` into a list of stings.   
   - Print the list and check the contents of the list of stings.

3. Modify the salutation to include the name of an invitee.
   - This is where the `replace()` function will come in handy.
   - This is ALSO where the `strip()` function can be used (extracting names from `names_of_invitees.txt`)

4. Modify the signature to include your name.

5. One cannot write a list to a file!
   - Therefore, convert the list into a **string**

6. Write the modified letter into a file. The file name should reflect who the addressee is.
   - For example, the file with the invite to *Katara* should be titled `to_Katara.txt`
   - You will have to generate the file name correctly,
   - Write the **modified** contents into the file, AND
   - <font color="red">Close the file</font> after writing.

6. Repeat this for all names in the file `names_of_invitees.txt`
