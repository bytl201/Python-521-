1. Check step: Read the file `starting_letter.txt` and print its contents.
   - Make sure you enter both file name and path correctly!
   - If you choose not to use the `with` keyword, make sure to <font color='red'>close</font> the file once you've read its contents.

````python
def read_starting_letter(file_path: str, ):
    # open the file in "read" mode
    with open(file_path, "r") as fobj:
        # read everything in 'fobj' in file
        file = fobj.read()
    # return 'file' as a str because 'fobj.read()' returns a str
    return file

# call function to return back str of the file contents
file_to_read = read_starting_letter("C:\\Users\\tlobd\\Desktop\\Python-521-\\birthday_party_invitation_problem_START\\input\\letters\\starting_letter.txt")
print(file_to_read)
````


2. Use the `readlines()` method associated with `File` objects to split the contents of `starting_letter.txt` into a list of stings.   
   - Print the list and check the contents of the list of stings.

````python
def read_file(file_path: str):
    # open file in 'read' mode
    with open(file_path, "r") as fobj:
        # iterate over each line in 'fobj'
        for row in fobj.readlines():
            # print each row and end print statement with ("") to not print extra white spaces when reading from file
            print(row, end="")

# print the file contents using readlines()
read_file("C:\\Users\\tlobd\\Desktop\\Python-521-\\birthday_party_invitation_problem_START\\input\\letters\\starting_letter.txt")
````

3. Modify the salutation to include the name of an invitee.
   - This is where the `replace()` function will come in handy.
   - This is ALSO where the `strip()` function can be used (extracting names from `names_of_invitees.txt`)

````python
def read_names(file_path: str):
    # array to hold names read from file
    name_list = []

    # open file in 'read' mode
    with open(file_path, "r") as fobj:
        # read each line in 'fobj'
        for row in fobj.readlines():
            # add each name to 'name_list' and strip() to get rid of extra whitespaces
            name_list.append(row.strip())
    # return all names from file
    return name_list

# returns the name from file
name_list = read_names("C:\\Users\\tlobd\\Desktop\\Python-521-\\birthday_party_invitation_problem_START\\input\\names\\names_of_invitees.txt")
````

4. Modify the signature to include your name.

````python
def write_letters(name_list: list):
    # read the file contents and hold it in 'file'. keep as a template to change for each name in 'name_list'
    file = read_starting_letter("C:\\Users\\tlobd\\Desktop\\Python-521-\\birthday_party_invitation_problem_START\\input\\letters\\starting_letter.txt")

    # iterate over each name in 'name_list'
    for i in name_list:

        # for robustness strip() each name again
        name = i.strip()

        # create a new file in correct location and write file name 
        with open(f"C:\\Users\\tlobd\\Desktop\\Python-521-\\birthday_party_invitation_problem_START\\output\\ready_to_send\\to_{name}.txt", "w") as fobj:

            # replace the 'invitee's name' and 'insert your name here' in 'file' variable and assign it to file_copy to then write to 'fobj'
            file_copy = file.replace("[invitee's name]", name).replace("[insert your name here]", "Tenzing Lobden")
            # write 'file_copy' contents to 'fobj'
            fobj.write(file_copy)

    # output to user to let them know the process is done.
    print(f"Sent letters to: {name_list}")

# call file to write letter to each person in 'name_list'
write_letters(name_list)
````
