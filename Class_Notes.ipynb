## Bubble sort

```python
list_rand = [9,8,7,6,5,4,3,2,1]

def bubbleSort(unordered_list: list):

    for i in range(len(unordered_list)):
        for j in range(0, len(unordered_list)-i-1):
            print(f"i = {unordered_list[i]}\nj = {unordered_list[j]}\nj+1 = {unordered_list[j+1]}\nlist = {unordered_list}\n")
            if unordered_list[j] > unordered_list[j+1]:
                unordered_list[j], unordered_list[j+1] = unordered_list[j+1], unordered_list[j] 

    return unordered_list
order_list = bubbleSort(list_rand)
print(order_list)
```

## Compute median of a list 

```python
def compute_median(numbers: list):
    numbers = sorted(numbers)

    length_list = len(numbers)

    if length_list % 2 != 0:
        return numbers[int(((length_list-1)/2))]
    return (numbers[int(length_list/2)] + numbers[int(length_list/2) - 1])/2

print(compute_median([1,2,3,4,5]))

print(compute_median([1,3,4,7]))
```

## Quick Sort

```python
def quick_sort(my_list: list):
    if len(my_list) <= 1:
        return my_list
    
    right_most_num = my_list.pop()

    smaller_right_num = [i for i in my_list if i < right_most_num]
    larger_right_num = [i for i in my_list if i >= right_most_num]

    return quick_sort(smaller_right_num) + [right_most_num] + quick_sort(larger_right_num)

quick_sort([5,3,7,1,9])
```
