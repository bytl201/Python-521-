```python
class LinearEquation:

    def __init__(self, a, b, c, d, e, f):
        '''
        Constructor for 'LinearEquation' Class, creates instance variable (a, b, c, d, e, f)
        '''
        self.a = a
        self.b = b
        self.c = c
        self.d = d
        self.e = e
        self.f = f
    
    def getX(self):
        '''
        Calculates 'X' and returns number
        '''
        num = (self.d * self.e) - (self.b * self.f)
        den = (self.a * self.d) - (self.b * self.c)

        return num/den

    def getY(self):
        '''
        Calculates 'Y' and returns number
        '''
        num = (self.a * self.f) - (self.e * self.c)
        den = (self.a * self.d) - (self.b * self.c)

        return num/den

    def isSolvable(self):
        '''
        Determine if equation is solvable and returns boolean
        '''
        return ((self.a * self.d) - (self.b * self.c)) != 0

# create instance of Class LinearEquation    
equation = LinearEquation(2, -3, 3, 4, 1, -1)

# output 
print(f"Solvable: {equation.isSolvable()}")

# format sig fig numbers
print(f"x = {equation.getX():.5f}")
print(f"y = {equation.getY():.5f}")
```
