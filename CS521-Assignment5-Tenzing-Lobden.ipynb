## 3. Palindromic primes

```python
def check_for_prime(n):
    """
    Description: Takes in a number and checks if prime or not
    """

    # if 'n' is less than 2, it's not prime
    if n < 2:
        return False
    
    # iterate from 2 to square root of 'n' to save computing power after square root of 'n' because we already iterated over it
    for i in range(2, int(n**0.5)+1):
        # check if divisible by numbers up to square root of n
        if n % i == 0:
            return False
    # return True if no number found divisible up to square root of n
    return True

def check_for_palindrome(n):
    """
    Description: Checks if the number is the same backwards.
    """

    # reverse the number and see if it's the same as the orginal number. Return the boolean value
    return str(n) == str(n)[::-1]
    
def palindromic_prime(n):
    """
    Description: Prints out Palindromic Prime numbers up to 'n'.
    """

    # variable to keep track of how many palindromic prime found to then print out on a new line
    count = 0
    # start at 2 because that's the first palindromic prime and save compute power
    index = 2


    # iterate until count is same as n or greater
    while count < n:

        # check if both boolean returns True
        if check_for_palindrome(index) and check_for_prime(index):

            # print out and tab 
            print(f"{index}", end = "\t")
            # add one to 'count'
            count+= 1

            # if there is 10 numbers on each line, then print() new line
            if count % 10 == 0:
                print()
        # move onto next number
        index +=1
palindromic_prime(14)
```
